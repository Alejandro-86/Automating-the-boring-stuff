### Open and Read PDF files

```
pdfFile = open('meetingminutes2.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFile)
print(pdfReader.numPages)
page = pdfReader.getPage(0)
page.extractText()
```

```
pdfReader = PyPDF2.PdfFileReader(open('encrypted.pdf', 'rb'))
pdfReader.isEncrypted
```

### Decrypt PDF files

```
pdfReader.decrypt('rosebud')
pdfReader.isEncrypted
pageObj = pdfReader.getPage(0)
pageObj.extractText()
```

### Combine PDF files

```
pdf1 = open('meetingminutes.pdf', 'rb')
pdf2 = open('meetingminutes2.pdf', 'rb')
reader1 = PyPDF2.PdfFileReader(pdf1)
reader2 = PyPDF2.PdfFileReader(pdf2)
writer = PyPDF2.PdfFileWriter()

for pagenum in range(reader1.numPages):
  pageObj = reader1.getPage(pagenum)
  writer.addPage(pageObj)
  

for pagenum in range(reader2.numPages):
  pageObj = reader2.getPage(pagenum)
  writer.addPage(pageObj)
  

Ofile = open('combinedminutes.pdf', 'wb')
writer.write(Ofile)
Ofile.close()
pdf1.close()
pdf2.close()
```

### Rotating pages

```
pdf = open('meetingminutes.pdf', 'rb')
reader = PyPDF2.PdfFileReader(pdf)
page = reader.getPage(0)
page.rotateClockwise(90)
```

```
writer = PyPDF2.PdfFileWriter()
writer.addPage(page)
oFile = open('rotatedPage.pdf', 'wb')
writer.write(oFile)
oFile.close()
pdf.close()
```

### Overlaying Pages

```
sourceF = open('meetingminutes.pdf', 'rb')
reader = PyPDF2.PdfFileReader(sourceF)
source1page = reader.getPage(0)
watermark = PyPDF2.PdfFileReader(open('watermark.pdf', 'rb'))
source1page.mergePage(watermark.getPage(0))
writer = PyPDF2.PdfFileWriter()
writer.addPage(source1page)
```

```
for pageNum in range(1, reader.numPages):
  page = pdfReader.getPage(pageNum)
  writer.addPage(page)

Ofile = open('watermarkedCover.pdf', 'wb')
writer.write(Ofile)
sourceF.close()
Ofile.close()
```

### Encrypting PDFs

```
file = open('meetingminutes.pdf', 'rb')
reader = PyPDF2.PdfFileReader(file)
writer = PyPDF2.PdfFileWriter()
for page in range(reader.numPages):
  writer.addPage(reader.getPage(page))
  
writer.encrypt('swordfish')
resultPdf = open('encryptedminutes.pdf', 'wb')
writer.write(resultPdf)
resultPdf.close()
```

```
def getText(filename):
  doc = docx.Document(filename)
  fullText = []
  for para in doc.paragraphs:
    fullText.append(' ' + para.text)
  return '\n\n'.join(fullText)
```

```
print(getText('demo.docx'))
```

## Adding Headings

```
doc = docx.Document()
doc.add_paragraph('Hello world')
para1 = doc.add_paragraph('This is a second paragraph. ')
doc.add_paragraph('This is yet another paragraph.')
para1.add_run('This text is being added to the second paragraph.')

doc.save('multipleParagraphs.docx')
```

```
doc = docx.Document()
for i in range(4):
  doc.add_heading('Header '+ str(i), i)
doc.save('headings.docx')
```

## Adding Line and Page Breaks

```
doc = docx.Document()
doc.add_paragraph('This is on the first page!')
doc.paragraphs[0].runs[0].add_break(docx.enum.text.WD_BREAK.PAGE)
doc.add_paragraph('This is on the second page!')
doc.save('twoPage.docx')
```

## Adding pictures

```
doc.add_picture('zophie.png', width=docx.shared.Inches(1), height=docx.shared.Cm(4))
doc.save('image.docx')
```
